```r
dataframe <- data.frame(
    id =c(1, 2, 3, 4, 5),
    name= c("Peter", "Amy", "Ryan", "Gary", "Michelle"),
    salary= c(623.30, 515.20, 611.00, 729.00, 843.25))

new_col <- cbind(dataframe, Department =c("finance", "human resources", "management", "accounting", "sales"))
new_col
```

```r
dataframe <- data.frame(
    id =c(1, 2, 3, 4, 5),
    name= c("Peter", "Amy", "Ryan", "Gary", "Michelle"),
    salary= c(623.30, 515.20, 611.00, 729.00, 843.25))

new_col <- cbind(dataframe, Department =c("finance", "human resources", "management", "accounting", "sales"))

barplot(new_col$salary[c(1,4,5)], names.arg = new_col$name[c(1,4,5)], ylab = 'Salary', main = 'Peter, Gary, and Michelle Salaries', col = "green", density = 25)
```

```r
a = median(new_col$salary)
b = max(new_col$salary)
c = min(new_col$salary)

x <- c(a, b, c)
mylabels <- c("median salary ($623.30)", "maximum salary($843.25)", "minimum salary($515.20)")

pie(x, label= mylabels, main= "Salary Analysis")
```
